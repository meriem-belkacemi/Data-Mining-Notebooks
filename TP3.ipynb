```python
def discretisation_methode2 (dataset, attribut) :
    N = dataset.shape[0]
    K = round(1 + (10/3)* math.log(N,10))
    print(K)
    largeur = (dataset[attribut].max()-dataset[attribut].min())/K
    
    class_intervals = [dataset[attribut].min() + largeur * i for i in range (K) ]
    class_intervals.append(dataset[attribut].max())
    print(class_intervals)
    
    dataset_copy = dataset.copy()
    for j in range(N) :
        for i in range (K) :
            if (class_intervals[i] <= dataset[attribut][j]) and (dataset[attribut][j] <= class_intervals[i+1]) :
                dataset_copy[attribut][j] = (class_intervals[i+1] + class_intervals[i])/2
    return dataset_copy
            
        
    

```

```python
dataset_copy = discretisation_methode2(dataset, "petal width")
dataset_copy["petal width"].values
```

```python
def normalize_min_max(dataset, attribut, new_min=0, new_max=1):

    min_old = dataset[attribut].min()
    max_old = dataset[attribut].max()

    dataset_normalized = dataset.copy()
    dataset_normalized[f'{attribut}_normalized'] = (dataset[attribut] - min_old) / (max_old - min_old) * (new_max - new_min) + new_min

    return dataset_normalized
```

```python
normalized_dataset = normalize_min_max(dataset, "petal length")
normalized_dataset
```

```python
def normalize_z_score(dataset, attribut):
    mean_old = dataset[attribut].mean()
    std_old = dataset[attribut].std()
    
    dataset_normalized = dataset.copy()
    dataset_normalized[f'{attribut}_z_score'] = (dataset[attribut] - mean_old) / std_old

    return dataset_normalized
```

```python
normalized_dataset = normalize_z_score(dataset, "petal length")
normalized_dataset
```

```python
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_std(data, mean):
    n = len(data)
    squared_deviations = [(x - mean) ** 2 for x in data]
    variance = sum(squared_deviations) / n
    std_deviation = variance ** 0.5
    return std_deviation

def normalize_z_score2(dataset, attribut):
    
    mean_old = calculate_mean(dataset[attribut])
    std_old = calculate_std(dataset[attribut], mean_old)
    
    dataset_normalized = dataset.copy()
    dataset_normalized[f'{attribut}_z_score'] = (dataset[attribut] - mean_old) / std_old

    return dataset_normalized
```

```python
normalized_dataset = normalize_z_score2(dataset, "petal length")
normalized_dataset
```

```python
import pandas as pd
def load_dataset (file):
    dataset = pd.read_excel(file)
    return dataset

path = "Exo1.xlsx"
dataset = load_dataset(path)
```

```python
def visualize_data (dataset) :
    print(dataset.head())
    print ("Nombre de lignes du dataset :", dataset.shape[0])
    print("Nombre de colonnes du dataset : ", dataset.shape[1])
    print("Noms des colonnes : ", dataset.columns.tolist())

visualize_data(dataset)
    
```

```python
def diagramme_dispersion(dataset,x,y) :
    plt.scatter(dataset[x],dataset[y])
    plt.title('Diagramme de dispersion')
    plt.xlabel('Valeurs de'+x)
    plt.ylabel('Valeur de'+y)
    plt.show()
```

```python
def histogramme(dataset,x) :
    plt.hist(dataset[x], bins=range(int(min(dataset[x])), int(max(dataset[x]))+1), edgecolor='black')
    plt.title('Histogramme de l attribut '+x)
    plt.show()
    
```

```python
categorical_attribute = dataset['Class']

category_counts = categorical_attribute.value_counts()

categories = category_counts.index
counts = category_counts.values

plt.figure(figsize=(8, 6))
plt.bar(categories, counts, color='skyblue')
plt.title('Distribution of Categorical Attribute')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
histogramme(dataset,"petal width")
```

```python
def remove_outliers (dataset, attribute) :
    Q1 = dataset[attribute].quantile(0.25)
    Q3 = dataset[attribute].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_removed_dataset = dataset[(dataset[attribute] >= lower_bound) & (dataset[attribute] <= upper_bound)]
    return outliers_removed_dataset
```

```python
outliers_removed_dataset = remove_outliers (dataset,"petal width")
```
